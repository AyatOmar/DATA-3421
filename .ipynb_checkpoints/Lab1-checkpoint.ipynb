## Data Transformation Methods

The Data Transformation methods we used was Mean Imputation, replacing the null values in the 'Sales Revenue_M' with the mean.
One Hot encoding was also used to convert the categorical variables in 'Genre' to their own columns containing 1s and 0s where, 1 is the variable.
